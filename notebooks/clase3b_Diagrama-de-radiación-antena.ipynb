# Clase 3b: Diagrama de radiación

```python
# Recuerda, utilizaremos arrays y pintaremos gráficas.
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
def espiraCuadrada(a=1,i=1,k=1):
    phi = np.linspace(0,2*np.pi, 200)
    N = 1j*k*i*a**2*np.sin(phi)
    plt.polar(phi, np.abs(N))
```

```python
espiraCuadrada()
```

```python
def dipolo(D=0.1):
    c=299792458
    f = 1e7
    landa=c/f
    H=landa*D
    i=1
    omega= f* 2 * np.pi
    mu=1.257e-6
    epsilon=8.85e-12
    k=omega*np.sqrt(mu*epsilon)
    phi = np.linspace(0,2*np.pi, 360)
    k_z = k*np.cos(phi)
    N = 2*k*i*( (np.cos(k_z*H) - np.cos(k*H) )/( k**2 - k_z**2) )
    plt.polar(phi, np.abs(N))
```

```python
interact(dipolo,D=(0.1,1))
```
